```python
import pandas as pd 
import numpy as np 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
import os 

base_dir = '/Users/woneyhoney/Desktop/weight/Again/Data_Analysis'
excel_file = '안전지수_클러스터(k=4).xlsx'
excel_dir = os.path.join(base_dir,excel_file)

df = pd.read_excel(excel_dir, sheet_name='Sheet1', usecols='C, B') # cluster_id, safe_score
# 데이터 시각화
sns.lmplot('cluster_id', 'safe_score', data=df, fit_reg = False, scatter_kws={"s":200}, hue="cluster_id")
plt.title('safe_level(0~3)')
plt.xlabel('cluster_id')
plt.ylabel('safe_score')
```
