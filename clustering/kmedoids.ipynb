```python
import pandas as pd 
import numpy as np 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
import os 

base_dir = '/Users/woneyhoney/Desktop/weight/Again/Data_Analysis'
excel_file = 'safety_level_3.xlsx'
excel_dir = os.path.join(base_dir,excel_file)

df = pd.read_excel(excel_dir, sheet_name='Sheet1', usecols='B, C') # x_latitude, y_longitude 

# visualize data point 
sns.lmplot('x_latitude', 'y_longitude', data=df, fit_reg = False, scatter_kws={"s":200})

# title 
plt.title('distribution of safety_level_3')

# x-axis label 
plt.xlabel('x_latitude')

# y-axis label
plt.ylabel('y_longitude')
```

```python
excel_file = 'safety_level_1.xlsx'
excel_dir = os.path.join(base_dir,excel_file)

df = pd.read_excel(excel_dir, sheet_name='Sheet1', usecols='B, C') # x_latitude, y_longitude 

# visualize data point 
sns.lmplot('x_latitude', 'y_longitude', data=df, fit_reg = False, scatter_kws={"s":200})

# title 
plt.title('distribution of safety_level_1')

# x-axis label 
plt.xlabel('x_latitude')

# y-axis label
plt.ylabel('y_longitude')
```

```python
# Running K-Medoids with a range of k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df)
    distortions.append(kmeanModel.inertia_)
```

```python
# Plotting the distortions of K-Means
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.utils import read_sample

# load list of points for cluster analysis
sample = read_sample('/Users/woneyhoney/Desktop/weight/Again/Data_Analysis/safety_level_3.txt')
# set random initial medoids
initial_medoids = [1, ]
# create instance of K-Medoids algorithm
kmedoids_instance = kmedoids(sample, initial_medoids)
# run cluster analysis and obtain results
kmedoids_instance.process();
clusters = kmedoids_instance.get_clusters()
# show allocated clusters
print(clusters)
```
