```python
import pandas as pd 
import numpy as np 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
import os 

base_dir = '/Users/woneyhoney/Desktop/weight/Again/Data_Analysis'
excel_file = 'safety_level_1.xlsx'
excel_dir = os.path.join(base_dir,excel_file)

df = pd.read_excel(excel_dir, sheet_name='Sheet1', usecols='B, C') # 위도, 경도 

data_points = df.values

kmeans = KMeans(n_clusters=300).fit(data_points) # 300개의 클러스터
df['cluster_id'] = kmeans.labels_

pd.set_option('display.max_rows', 300)
df['cluster_id'].value_counts() < 545
```

```python
df_cpy_1 = df # 원본 복사본

# True인거만 선택 - 노가다 부분
df = df[df['cluster_id'] != 95]
df = df[df['cluster_id'] != 93]
df = df[df['cluster_id'] != 27]
df = df[df['cluster_id'] != 54]
df = df[df['cluster_id'] != 48]
df = df[df['cluster_id'] != 80]
df = df[df['cluster_id'] != 96]
df = df[df['cluster_id'] != 97]
df = df[df['cluster_id'] != 66]
df = df[df['cluster_id'] != 84]

# 열 날리기 
df.drop('cluster_id', axis=1)
```
